```python
# ensure packages are installed in the notebook environment before importing
%pip install -q fastai
%pip install -q fastdownload
%pip install -q ddgs

from ddgs import DDGS
from fastcore.all import *
from fastdownload import download_url
from fastai.vision.all import *

def search_images(term, max_images=30):
    print(f"Searching for '{term}'")
    with DDGS() as ddgs:
        results = ddgs.images(term, max_results=max_images)
        return [r["image"] for r in results]
```

```python
# attempt multiple search results until we successfully download and open an image
urls = search_images('bullish pattern', max_images=10)
for i, url in enumerate(urls):
	try:
		print(f"Downloading image {i+1}/{len(urls)}")
		download_url(url, 'bullish.jpg', show_progress=False)
		# verify the image can be opened; reopen after verify
		img = Image.open('bullish.jpg')
		img.verify()
		img = Image.open('bullish.jpg')
		display(img.to_thumb(256,256))
		break
	except Exception as e:
		print(f"Failed to use url ({url}): {e}\nTrying next...")
else:
	raise RuntimeError("Could not download a valid 'bullish' image from search results")
```

```python
urls = search_images('bearish pattern', max_images=10)
for i, url in enumerate(urls):
	try:
		print(f"Downloading image {i+1}/{len(urls)}")
		download_url(url, 'bearish.jpg', show_progress=False)
		img = Image.open('bearish.jpg')
		img.verify()
		img = Image.open('bearish.jpg')
		display(img.to_thumb(256,256))
		break
	except Exception as e:
		print(f"Failed to use url ({url}): {e}\nTrying next...")
else:
	raise RuntimeError("Could not download a valid 'bearish' image from search results")
```

```python
searches = 'bullish','bearish'
path = Path('bullish_or_bearish')
from time import sleep

for o in searches:
    dest = (path/o)
    dest.mkdir(exist_ok=True, parents=True)
    download_images(dest, urls=search_images(f'{o} pattern'))
    resize_images(path/o, max_size=400, dest=path/o)
```

```python
failed = verify_images(get_image_files(path))
failed.map(Path.unlink)
len(failed)
```

```python
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[Resize(192, method='squish')]
).dataloaders(path, bs=32)

dls.show_batch(max_n=6)
```

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(3)
```

```python
sample_photo = PILImage.create('bullish.jpg')
sample_photo.show()
pattern_type,_,probs = learn.predict(sample_photo)

print(f"This is a: {pattern_type} pattern.")
print(f"Probability it's a bullish pattern: {probs[0]:.4f}")
print(f"Probability it's a bearish pattern: {probs[1]:.4f}")
```

```python
sample_photo = PILImage.create('bearish.jpg')
sample_photo.show()
pattern_type,_,probs = learn.predict(sample_photo)

print(f"This is a: {pattern_type} pattern.")
print(f"Probability it's a bullish pattern: {probs[0]:.4f}")
print(f"Probability it's a bearish pattern: {probs[1]:.4f}")
```
